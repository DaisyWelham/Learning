```python
import numpy as np

class DataLoader:
    def __init__(self, savename):
        self.savename = savename
        self.inputsMatrix = np.load("inputsMatrix" + savename + ".npy")
        self.outputsMatrix = np.load("outputsMatrix" + savename + ".npy")
        self.dataMatrix = np.load("dataMatrix" + savename + ".npy")

    def printPercentiles(self, matrix, label, names):
        nVariables = matrix.shape[1]
        for i in range(nVariables):
            data = matrix[:, i, :]
            p0 = np.percentile(data, 0, axis=0)
            p25 = np.percentile(data, 25, axis=0)
            p50 = np.percentile(data, 50, axis=0)
            p75 = np.percentile(data, 75, axis=0)
            p100 = np.percentile(data, 100, axis=0)
            std = np.std(data, axis=0)

            print(f"{label} {names[i]}:")
            print(f"  0th:   {p0[-1]}")
            print(f"  25th:  {p25[-1]}")
            print(f"  50th:  {p50[-1]}")
            print(f"  75th:  {p75[-1]}")
            print(f"  100th: {p100[-1]}")
            print(f"  STD:   {std[-1]}")
            print("")

    def processAll(self):
        inputNames = ["inlet level", "well level", "storage level"]
        outputNames = ["pump 1 activation", "pump 2 activation", "valve state"]
        dataNames = ["overflow rate", "pump use"]

        print("Inputs:")
        self.printPercentiles(self.inputsMatrix, "Input", inputNames)
        print("Outputs:")
        self.printPercentiles(self.outputsMatrix, "Output", outputNames)
        print("Data:")
        self.printPercentiles(self.dataMatrix, "Data", dataNames)

savenames = [
    "random", 
    "plc", 
    "actorcritic", 
    "imitationactorcritic", 
    "finetunedimitationactorcritic", 
    "deepq", 
    "imitationq", 
    "finetunedimitationq"
]

for name in savenames:
    print("========================")
    print(f"Results for {name}:")
    loader = DataLoader(name)
    loader.processAll()
```
