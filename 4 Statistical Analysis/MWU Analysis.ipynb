```python
import numpy as np
import pandas as pd
from scipy.stats import mannwhitneyu

agent_names = [
    "random", 
    "plc", 
    "actorcritic", 
    "imitationactorcritic", 
    "finetunedimitationactorcritic", 
    "deepq", 
    "imitationq", 
    "finetunedimitationq"
]
overflow_data = {}
for name in agent_names:
    matrix = np.load(f"dataMatrix{name}.npy")
    overflow_values = matrix[:, 0, :].flatten()  # Adjust axis as needed
    overflow_data[name] = overflow_values

results = []
for i, agent1 in enumerate(agent_names):
    for agent2 in agent_names[i + 1:]:
        u_stat, p_value = mannwhitneyu(overflow_data[agent1], overflow_data[agent2], alternative='two-sided')
        results.append({
            'Agent 1': agent1,
            'Agent 2': agent2,
            'U statistic': u_stat,
            'p-value': p_value
        })

df = pd.DataFrame(results)
print(df.to_string(index=False))
```
